```python
columns = games.columns.tolist()
columns = [c for c in columns if c not in ["bayes_average_rating", "average_rating", "type", "name", "id"]]
```

```python
columns
```

```python
target = "average_rating"
```

```python
train = games.sample(frac=0.8,random_state=1)
test = games.loc[~games.index.isin(train.index)]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
LR_model = LinearRegression()
LR_model.fit(train[columns], train[target])
```

```python
predictions = LR_model.predict(test[columns])
mean_squared_error(predictions, test[target])
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RFR_model = RandomForestRegressor(n_estimators=100, min_samples_leaf=10, random_state=2)
RFR_model.fit(train[columns],train[target])
```

```python
predictions = RFR_model.predict(test[columns])
mean_squared_error(predictions, test[target])
```

```python
rating_LR = LR_model.predict(test[columns].iloc[0].values.reshape(1,-1))
rating_RFR = RFR_model.predict(test[columns].iloc[0].values.reshape(1,-1))
print(rating_LR)
print(rating_RFR)
print(test[target].iloc[0])
```
